```python
f = open("bai2.inp", "r")
input = f.read().strip().split(sep='\n')
f.close()
input = [s.strip().split() for s in input]
print(input)
```

```python
f = open("bai2.out", "r")
output = f.read().strip().split()
f.close()
print(output)
```

```python
def isAccepted(input, output):
  # Trường hợp -1
  if output[0] == "-1":
    import networkx as nx
    graph = nx.Graph()
    graph.add_edges_from(input[1:])
    graph.add_nodes_from(input[0][2:])
    if nx.has_path(graph, input[0][2], input[0][3]):
      return False
    else:
      return True
  # Trường hợp khác -1    
  if output[0] != input[0][-2] or output[-1] != input[0][-1]:
    return False
  for i in range(len(output) - 1):
    if [output[i], output[i+1]] not in input and [output[i+1], output[i]] not in input:
      return False
  return True
```

```python
if isAccepted(input, output):
  print("Correct!! ^o^")
else:
  print("Incorrect!! X_X")
```
