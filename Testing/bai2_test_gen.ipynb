```python
import networkx as nx
from random import randint

graph = nx.gnp_random_graph(n=randint(2,13), p=0.5, directed=False)
n = graph.number_of_nodes()
m = graph.number_of_edges()
s = randint(0,n-1)
d = randint(0,n-1)
while d == s:
  d = randint(0,n-1)  
```

```python
f = open("bai2.inp", "w")
f.write("")
f.close()
f = open("bai2.inp", "a")
```

```python
f.write(str(n) + ' ' + str(m) + ' ' + str(s) + ' ' + str(d) + '\n')
for edge in graph.edges:
  f.write(str(edge[0]) + ' ' + str(edge[1]) + '\n')
```

```python
f.close()
f = open("bai2.inp", 'r')
print(f.read())
nx.draw(graph, with_labels=True, font_weight='bold')
```
