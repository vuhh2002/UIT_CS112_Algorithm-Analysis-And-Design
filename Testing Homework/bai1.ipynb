```python
# Hàm sinh số và mảng ngẫu nhiên
import random

print(random.randint(1,200))
print(random.randint(0,7))
print(random.randint(2,3))
```

```python
# Hàm sinh mảng ngẫu nhiên
import random

def generateRandomIntegerList(length, integerLowerBound, integerUpperBound):
  return [random.randint(integerLowerBound, integerUpperBound) for i in range(length)]

print(generateRandomIntegerList(4,1,200))
print(generateRandomIntegerList(5,0,7))
print(generateRandomIntegerList(6,2,3))
```

```python
# Hàm sinh chuỗi ngẫu nhiên (với ký tự trong chuỗi từ 'a' đến 'z') 
import random
import string

def generateRandomString(length):
    letter = string.ascii_lowercase
    randomString = ''.join(random.choice(letter) for i in range(length))
    return randomString

print(generateRandomString(8))
print(generateRandomString(60))
print(generateRandomString(4))
```

```python
# Hàm sinh cây không trọng số
import networkx as nx

tree = nx.random_tree(n=10)

nx.draw(tree, with_labels=True, font_weight='bold')
print(nx.forest_str(tree, sources=[0]))
```
